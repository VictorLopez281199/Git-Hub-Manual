# 🚀 Guía Interactiva de GitHub

¡Bienvenido a esta guía visual e interactiva de GitHub! Aquí aprenderás a sacar el máximo provecho a las herramientas que ofrece GitHub para desarrolladores, equipos y entusiastas del código.

## 🐛 Issues: Reporte y Seguimiento de Problemas

<font color='blue'>**¿Qué son los Issues?**</font><br>
Son tickets donde puedes:
- Reportar errores 🐞
- Proponer mejoras ✨
- Hacer preguntas ❓

🔧 **Cómo crear un Issue:**
1. Ve a la pestaña <kbd>Issues</kbd>
2. Haz clic en <kbd>New issue</kbd>
3. Escribe un título y descripción clara 📝
4. Etiqueta, asigna y organiza si es necesario 🏷️

## 🔀 Pull Requests: Propón y Revisa Cambios

<font color='green'>**¿Qué es un Pull Request?**</font><br>
Es una solicitud para fusionar tus cambios en una rama del repositorio.

🛠️ **Pasos para usarlo:**
1. Crea una rama 📂
2. Haz tus cambios 💡
3. Ve a <kbd>Pull Requests</kbd> > <kbd>New Pull Request</kbd>
4. Describe tu cambio y solicita revisión 👀

## 📋 GitHub Projects: Gestión Visual de Tareas

<font color='orange'>Organiza tu trabajo como un profesional</font> 🔧

✨ **Usa tableros estilo Kanban:**
- "To Do" ✅
- "In Progress" 🔄
- "Done" 🎉

🎯 Ideal para seguimiento ágil de tareas, bugs y características.

## 🤖 Automatización con GitHub Actions

Automatiza flujos de trabajo: tests, builds, despliegues.

⚙️ **Ejemplo básico:**
```yaml
name: CI
on: [push]
jobs:
  build:
    runs-on: ubuntu-latest
    steps:
    - uses: actions/checkout@v2
    - name: Setup Python
      uses: actions/setup-python@v2
      with:
        python-version: '3.9'
    - name: Install dependencies
      run: pip install -r requirements.txt
```
🌐 Aprende más en [docs.github.com/actions](https://docs.github.com/actions)

## ✂️ GitHub Gists: Comparte Código Fácilmente

Perfecto para compartir:
- Snippets ✍️
- Scripts 🔐
- Notas rápidas 🧠

📌 Visita: [gist.github.com](https://gist.github.com) y crea el tuyo con facilidad.

## 🌐 GitHub Pages: Publica tu Sitio Web

¡Crea tu portafolio o documentación con GitHub Pages!

🚀 **Pasos:**
1. Ve a <kbd>Settings</kbd> > <kbd>Pages</kbd>
2. Selecciona una rama 📂
3. ¡Tu sitio estará en línea! 🌍

🖼️ Soporta HTML, Markdown y Jekyll.

## 💻 GitHub Codespaces: IDE en la Nube

Edita, compila y ejecuta código directamente desde tu navegador.

🌟 **Ventajas:**
- Rápido
- Basado en VS Code
- Ideal para colaboración

🔧 Haz clic en <kbd>Code > Codespaces > Create codespace</kbd>

## 🧠 GitHub.dev: Edición Instantánea

Cambia `github.com` por `github.dev` en la URL para editar tu repositorio sin instalar nada.

💡 Ejemplo:
```
https://github.dev/usuario/repositorio
```
🔎 Ideal para cambios rápidos y exploración del código.
